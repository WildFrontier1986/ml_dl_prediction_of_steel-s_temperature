```python
# Создание функции, формирующей и выводящей
# датафреймы с информацией о них
# 1. Объявление функции df_create():
def df_create(data, name, count):
    # 2. Создание датафрейма с проверкой ошибок
    try:
        df_a = pd.read_csv(f'/datasets/final_steel/{data}')
    except:
        df_a = pd.read_csv(Path(pathlib.Path.cwd(),
                                f'datasets/final_steel/{data}'))
    # 3. Вывод информации
    print(f'первые {count} строк с данными {name}'.upper())
    display(df_a.head(count))
    print('общая информация'.upper())
    print()
    display(df_a.info())
    print()
    print('список характерных значений'.upper())
    display(df_a.describe())
    print('Количество явных дубликатов'.upper())
    display(df_a.duplicated().sum())
    return df_a
```

```python
# Создание функции, выводящей тепловую карту
# коэффициентов корреляции
# 1. Объявление функции heat_crr()
def heat_corr(data, method_input):
    # 2. Формирование маски для нижнего треугольника
    # матрицы коэффициентов корреляции
    mask = np.triu(np.ones_like(data.corr(method=method_input), dtype=bool))
    # 3. Формирование пользовательской палитры
    cmap1 = sns.color_palette('hls', data.shape[1])
    # 4. Задание размеров графика
    plt.figure(figsize=(20, 20),
               facecolor='lightgray')
    # 5. Вывод тепловой карты коэффициентов корреляции
    sns.heatmap(
        np.round(data.corr(method=method_input), decimals=2),
        mask=mask,
        cmap=cmap1,
        vmax=1,
        vmin=-1,
        center=0,
        annot=True,
        square=True,
        linewidths=.45,
        cbar_kws={'shrink': .6,
                  'orientation': 'vertical'}
    )
    plt.title(f'тепловая карта коэффициентов корреляции {method_input}'.upper())
plt.show()
```

### Исследование модели полносвязных нейронных сетей - Dense

Необходимо учитывать склонность моделей нейронных сетей к переобучению.

```python
# Формирование объекта структуры данных StandardScaler
scaler_nn = StandardScaler()
```

```python
# Запуск на обучающей выборке
scaler_nn.fit(features_train)
```

```python
# Масштабирование данных обучающей выборки
# и вывод на экран первых 5 строк
features_train_scaled = scaler_nn.transform(features_train)
```

```python
# Масштабирование данных тестовой выборки
# и вывод на экран первых 5 строк
features_test_scaled = scaler_nn.transform(features_test)
```

```python
# Создание функции для подбора лучших параметров
# с помощью тюнера для обучения модели
# 1. Объявление функции build_neural()
def build_neural(hp):
    # 2.1. Создание структуры модели и дополнение слоями
    # с перебором параметров слоёв
    model = Sequential()
    # 2.2. Цикл перебора количества скрытых слоёв и их параметров
    for j in range(hp.Int('num_layers', 1, 3)):
        model.add(Dense(
            units=hp.Int(f'units_{j}', min_value=16, max_value=64, step=8),
            activation=hp.Choice('activation', values=['linear', 'relu'])
        ))
    # 2.3. Создание переменной с вариациями коэффициента
    # регуляризации в выходном слое и добавление выходного слоя
    activ_reg = hp.Choice('activity_regularizer', values=[.0001, .0005, .001, .005, .01])
    model.add(Dense(
        units=1,
        activation=hp.Choice('activation_out', values=['linear', 'relu']),
        activity_regularizer=L1(l1=activ_reg)
    ))
    # 2.4. Создание переменной с вариациями коэффициента
    # скорости обучения и компилирование настроек модели
    l_rate = hp.Choice('learning_rate', values=[.001, .005, .01, .05, .1])
    model.compile(
        loss='mean_squared_error',
        optimizer=Adam(learning_rate=l_rate),
        metrics=['mae']
                 )
    # 3. Возвращение функции
    return model
```

```python
# Задание алгоритма подбора параметров
tuner = Hyperband(hypermodel=build_neural,
                  objective=['val_mae'],
                  max_epochs=25,
                  factor=2,
                  hyperband_iterations=1,
                  seed=3102022,
                  tune_new_entries=True,
                  allow_new_entries=True)
```

```python
# Вывод сводки гиперпараметров для поиска
tuner.search_space_summary()
```

```python
# Поиск лучших конфигураций гиперпараметров
tuner.search(features_train_scaled,
             target_train,
             batch_size=16,
             epochs=25,
             verbose=1,
             validation_split=.25,
             shuffle=True)
```

```python
# Формирование и вывод переменной с
# лучшей конфигурацией гиперпараметров
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
best_hps.values
```
